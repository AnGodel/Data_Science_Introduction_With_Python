## <a name="M1">1. Data Driven Decision Makining (DDDM)</a>

* La toma de decisiones basada en datos (DDDM) implica ***tomar decisiones respaldadas por los datos*** en lugar de tomar decisiones que sean intuitivas o basadas solo en observaciones.


* Los ***datos por si solos no son relevantes*** y por tanto debemos de trabajar con ellos para ***obtener un conocimiento del entorno*** en el que nos movemos y ***tomar decisiones en base al conocimiento y sabiduría*** obtenidos de los datos.


* La pirámide del dato al conocimiento sería la siguiente:

<img src="./imgs/00_02_ds.png" style="width: 400px;"/>
<span></span><br><br>

* Definimos estos 4 niveles de la siguiente manera:
<span></span><br><br>
    * ***Datos***: Lo podemos considerar como la unidad mínima semántica que por si solo es irrelevante para la toma de decisiones.
<span></span><br><br>
    * ***Información***: Puede definirse como un conjunto de datos procesados que tienen un cierto significado, y que por lo tanto son de utilidad para quién debe tomar decisiones.
<span></span><br><br>
    * ***Conocimiento***: Hace uso de la información para la toma de decisiones; es decir, da significado a la información.
<span></span><br><br>
    * ***Sabiduría***: Es el conocimiento profundo que se adquiere a través del conocimiento o de la experiencia.


* En los últimos años las empresas están ***invirtiendo mucho dinero en el conocido mundo de "Big Data"*** porque tienen la necesidad de ***tomar buenas decisiones para el bien de su negocio basado en el conocimiento extraído de los datos y no del conocimiento de "gurus"*** o personas que tienen un conocimiento importante del negocio en el que se mueven.

## <a name="M2">2. El Big Data</a>


* La ***necesidad de procesar "cuantos más datos mejor"*** para la ***obtención de conocimiento y toma de decisiones***, ha hecho que los sistemas de almacenamiento y procesado de datos tradicionales se hayan quedado pequeños, por ello han ***surgido nuevas tecnologías capaces de procesar grande volúmenes de datos***.


* Las necesidades del mercado hacen que cada año ***aumenten de manera significativa el número de tecnologías Big Data***:

### - Fases del Big Data


* El ***Big Data no es solo recopilar datos y almacenarlos, hay que tener la capacidad de procesarlos, extraer conocimiento de los datos y aplicarlo en la toma de decisiones***.


* El Big Data lo podemos descomponer en los siguientes pasos, aunque dependiendo de la bibliografía a consultar puede que algunos pares de fases los considere como una sola:

    1. ***Origen de los datos***.
    2. ***Integración*** de los datos.
    3. ***Almacenamiento*** de los datos.
    4. ***Procesado*** de los datos (Análisis de datos, Machine Learning, etc.)
    5. ***Visualización*** de la información.

## <a name="M3">3. ¿Que es la ciencia de Datos?</a>


* La ***ciencia de datos*** es un campo interdisciplinario que ***involucra métodos científicos, procesos y sistemas para extraer conocimiento o un mejor entendimiento de datos***, lo cual es una continuación de algunos campos de análisis de datos como la ***estadística, la minería de datos, el aprendizaje automático y la analítica predictiva***.


* También se define la ciencia de datos como "***Un concepto para unificar estadísticas, análisis de datos, aprendizaje automático y sus métodos relacionados para comprender y analizar los fenómenos reales***", empleando técnicas y teorías extraídas de muchos campos dentro del contexto de las matemáticas, la estadística, la ciencia de la información y la informática.

### - Big Data vs Data Science


* El ***Big Data y la Ciencia de Datos son dos campos que van de la mano*** ya que la ***proliferación del campo de la Ciencia de Datos se debe a dos factores***: uno, la ***evolución tecnológica en las computadoras*** que permiten tener cada vez más capacidad de computo y dos, las ***tecnologías Big Data*** desarrolladas que permiten el almacenamiento y procesamiento de datos de una manera distribuida y muy optimizada.


* ***Todd Yellin, vicepresidente de producto de Netflix***, comentaba: <span style="color:red">***“El big data, como me gusta decir, es una gran cantidad de basura que esconde diamantes. Entonces lo que hacen los algoritmos de machine learning es revelar y mostrar cuáles son esos diamantes. Cuál es la información importante..."***</span>. Con esta definición podemos entender la importante relación que hay entre el Big Data y la Ciencia de Datos.


## <a name="M4">4. Ciclo de vida de un proyecto Data Science</a>


* ***Los proyectos*** que se llevan a cabo en el mundo de la ***Ciencia del Dato***, al igual que el resto de proyectos de otras áreas de estudio ***tienen como objetivo resolver algún tipo de problema*** que esté causando perdidas a una empresa, asociación, etc o que quieran obtener un aumento de capital (económico, social, humano, etc.) en algún área especifica de la organización.


* Por tanto desde que ***se tiene el Dato hasta que se toma una decisión*** deben de realizarse una serie de pasos para llegar al objetivo final que es el de tomar una decisión que permita en definitivas cuentas aumentar el capital de la organización.

### 4.1. Entender Necesidades del Negocio


* Lo primero que hay que hacer en un proyecto Data Science es ***saber que caso de negocio o problema se quiere abordar***.


* No vale de nada empezar un proyecto si no se tiene una ***necesidad de negocio o un objetivo claro para abordar*** como por ejemplo optimizar coses, aumentar ventas, reducción del Churn, etc.


* Una vez entendida la necesidad se tienen que ***enumerar las fuentes de datos necesarias*** para ver si a priori se puede abordar el proyecto con garantías de éxito.


* ***Enumerar en exceso o en defecto las fuentes de datos necesarias puede llevar al fracaso del proyecto*** por exceso ***costes***, tiempos o por falta de ***calidad en el Insigth*** del producto.


### 4.2. Recolección de Datos


* La recolección de datos es el primer paso o la ***primera fase del Big Data***.

* Las ***fuentes de información*** pueden ser muy diversas y los ***datos pueden estar estructurados o no***:
    - Redes sociales
    - IoT
    - Logs
    - Webs

* Es importante saber ***obtener estas fuentes de datos*** via ***Syslog, APIs, scraping***, etc. ya que sin los datos no se puede abortar ningún tipo de proyecto Big Data


### 4.3. Limpieza y Organización de Datos


* Limpiar y organizar los datos es la ***tarea por excelencia*** en el negocio de los datos.


* A este tipo de tarea se le conoce como ***ETL*** (Extraer-Transformar-Cargar datos(Load)).


* En ciertas ocasiones es necesario el ***uso de tecnologías Big Data como Spark*** para realizar este tipo de tareas debido a los grandes volúmenes de datos que se maneja.


### 4.4. Análisis Exploratorio de Datos (EDA)


* Una vez que ya se tienen los datos limpios, organizados y estructurados entran en juego los ***analistas de datos (Business Intelligence)*** con el objetivo de ***entender el contenido de los datos***.


* El objetivo de esta fase es la de ***obtener información a partir de los datos en bruto***; es decir, empezar a ***entender el significado de los datos y obtener conocimiento*** de ellos.


* El proceso exploratorio puede volver a realizar una ***nueva limpieza de los datos, generar datos derivados, agregaciones de datos o generar estadísticas descriptivas***.


* La ***visualización de datos*** utilizando diferentes tipos de gráficos, mapas, etc. permiten obtener un ***mayor y mejor entendimiento de la información***.


### 4.5. Algoritmos de Aprendizaje: Modelos


* Los ***Algoritmos de Aprendizaje generan Modelos*** (que son una representación de la realidad) a partir de un conjunto de datos. Explotando estos modelos, nos permitirán tomar decisiones o acciones a realizar en nuestro caso de negocio.


* El área del ***(Deep) Machine Learning*** tiene como objetivo entrenar modelos (con Algoritmos de Aprendizaje) para que posteriormente sean capaces de ***realizar tareas de forma autónoma***:


* Algunos de los problemas que podemos resolver con la generación de estos modelos pueden ser:

    - Clasificación y estimación de probabilidades
    - Estimación o predicción de valores
    - Clustering
    - Patrones de similitud
    - Profiling
    - Recomendaciones
    - Reducción de datos
    - Causalidad



### 4.6. Evaluación de Modelo/s


* Los ***Modelos generados por los Algoritmos de Aprendizaje*** se pueden considerar como una representación de la realidad que ha sido aprendida de los datos.


* Estos ***Modelos hay que evaluarlos*** de tal manera que podamos ver en un entorno controlado ***como de buenas o malas son las decisiones o acciones que toma***.


* Una manera de hacerlo es ***reservarse un subconjunto de datos*** de los cuales conocemos la decisión o acción a tomar real y ***ver si coincide o no la decisión o acción tomada por el modelo respeto a la decisión o acción real***.


### 4.7. Puesta en Producción


* Todo proyecto relacionado con los datos para la posterior toma de decisiones no culmina si no hay ***un entregable de Insigths de forma más o menos recurrente*** para ***la posterior toma de decisiones***.


* Este tipo de productos puede ofrecer algún tipo de ***Dashboard para el entendimiento o monitorización de los datos de forma visual o como añadido algún tipo de predicción o recomendación para las tomas de decisiones***.
