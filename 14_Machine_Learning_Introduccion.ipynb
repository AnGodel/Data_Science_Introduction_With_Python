# <a name="M1">1. Inteligencia Artificial</a>


* Hoy en día existen muchas definiciones sobre Inteligencia Artificial, pero podríamos dar estas siguientes tres definiciones sobre IA:
<span></span><br><br>
    * **<span style="color:blue">Def 1</span>**: ***Cualquier técnica que permite a un ordenador imitar la inteligencia humana o natural, a través de lógica, reglas, Machine Learning, etc.***

    2. **<span style="color:red; font-size:18px">Deep Learning</span>**: Técnica de Machine Learning en la cual ***redes neuronales*** profundas ***aprenden de una gran cantidad de datos***.

# <a name="M2">2. Machine Learning</a>


## 2.1 ¿Qué es el Machine Learning?


* Veamos dos de las definiciones más conocidas del Machine Learning:

<img src="./imgs/13_02_ml.png" style="width: 600px;"/>

* El ***Machine Learning (ML) o Aprendizaje Autónomo*** es una ***área de la Inteligencia Artificial*** (IA) que tiene como ***objetivo crear sistemas capaces de aprender por ellos mismos*** a partir de un conjunto de datos (data set), ***sin ser programados de forma explícita***.


* Para que estos sistemas puedan aprender por ellos mismos, ***se utilizan*** una serie de ***técnicas y algoritmos capaces de crear modelos predictivos, patrones de comportamiento, etc***.


* En ML encaja todo proceso de resolución de ***problemas, basados más o menos explícitamente en una aplicación rigurosa de la teoría de la decisión estadística***; por tanto, es muy normal que el área del ML se solape con el área de la estadística


* En ML; a diferencia de la estadística, ***se centra en el estudio de la complejidad computacional de los problemas***, ya que gran parte de estos son de la clase NP-completo (o NP-hard) y por tanto el reto del ML está en diseñar soluciones factibles para este tipo de problemas


## 2.2 Tipos de Aprendizaje


* Dado que el ***Machine Lerning tiene como objetivo crear sistemas capaces de aprender por ellos mismos*** a partir de un conjunto de datos, ***sin ser programados de forma explícita***, tenemos que ver de que manera estos sistemas son capaces de aprender.


* Existen una serie de ***tipos de aprendizaje (muy similares a como aprendemos los humanos y los seres vivos)*** para que los sistemas sean capaces de aprender a partir de una experiencia.


* Estos tipos de aprendizaje son:
<span></span><br><br>
    - **<span style="color:red; font-size:18px">Aprendizaje supervisado</span>**: Tipo de aprendizaje en el que ***se tiene la información completa de los datos de entrenamiento***; es decir, ***los datos de entrada y la salida de los mismos***. Es el tipo de aprendizaje que mejores resultado ofrece ya que es el que más información tiene. Este aprendizaje resuelve los siguientes típos de problemas:
    
        + Regresión
        + Clasificación
<span></span><br><br>
    - **<span style="color:red; font-size:18px">Aprendizaje no supervisado</span>**: Tipo de aprendizaje en el que únicamente ***se disponen de los datos de entrada y tiene como objetivo el obtener información sobre la estructura del dominio de salida***. Este aprendizaje resuelve los siguientes típos de problemas:
    
        + Clustering (agrupación)
        + Reducción de la dimensionalidad
<span></span><br><br>
    - **<span style="color:red; font-size:18px">Aprendizaje por refuerzo</span>**: Es un tipo de aprendizaje en el que ***los sistemas aprenden de manera exploratoria, recibiendo recompensas o castigos en función de si las acciones tomadas por el sistema son buenas o malas***. Es un tipo de aprendizaje que se basa en el “argumentum ad baculum”, utilizado normalmente en la educación de los animales.


* Cada uno de estos ***tipos de aprendizaje*** tiene una serie de ***algoritmos de aprendizaje*** que a partir de unos ***datos de entrada*** devuelven un ***modelo*** (o hipótesis) con el que poder ***predecir*** nuevas ocurrencias.

### 2.2.1 Aprendizaje Supervisado


* Para que los ***algoritmos de aprendizaje supervisado aprendan***, se necesitan ***datos etiquetados***; es decir, que dado un estado (un partido de fútbol, una prueba diagnostica, una transación bancaria, etc) se conozca su resultado. Por tanto hay que conocer tanto el dominio de entrada como el dominio de salida.


* El ***proceso*** dentro del aprendizaje supervisado sería el siguiente, tanto ***para la resolución de problemas de Clasificación*** como de ***Regresión***:
<span></span><br><br>
    1. **<span style="color:red; font-size:18px">Datos de Entrenamiento</span>**: Se necesitan un conjunto de datos de entrenamiento donde:
        + Se conozcan el dominio o las ***varibles de entrada 'X'***.
        + Se conozca el dominio o el ***valor resultante 'Y'***,
            - Y (Regresión): Puede tener ***infinitos valores***.
            - Y (Clasificación): Tiene que tener ***'n' valores finitos***.
<span></span><br> <br>
    2. **<span style="color:red; font-size:18px">Algoritmos de Aprendizaje</span>**: El algoritmo de aprendizaje es el ***encargado de general un modelo (o hipótesis)*** que posteriormente sirva para predecir. Este modelo lo genera ***minimizando*** lo que se conoce como la **"Función de Perdida"** (o **"Función de Coste"**) que es la ***suma de los errores cometidos en la predicción de cada una de las observaciones del conjunto de datos de entrenamiento***.
<span></span><br> <br>     
        + Por tanto ***definimos la función de perdida*** como la suma de cada uno de los errores de cada observación de los datos de entrenamiento:    
<span style="font-size:20px">$$función\;perdida = E(parámetros) = \frac{1}{N} \sum_{i=1}^{N} error_{i}$$</span>
<span></span><br>
            - Siendo ***N***, el ***número total de observaciones***.
            - Siendo ***i***, el ***índice de cada observación***.
<span></span><br><br>
        + Podemos ***definir*** el ***error*** de una observación en un ***problema de Regresión*** como:     
<br><br><span style="font-size:18px">$$error(x,y,h(x)) = \frac{1}{2}(h(x) - y)^2$$</span>
<span></span><br><br>
        + Podemos ***definir*** el ***error*** de una observación en un ***problema de clasificación*** como:
<br><br><span style="font-size:18px">$$\begin{matrix}error(x,y,h(x)) = 1\;si\;h(x)\neq y\\ error(x,y,h(x)) = 0\;si\;h(x)= y\end{matrix}$$</span>
<span></span><br><br>
    3. **<span style="color:red; font-size:18px">Modelo</span>**: Es una **"representación de la realidad"** que el ***algoritmo de aprendizaje ha aprendido basandose en los datos de entrenamiento***. Un modelo (tambien se le conoce como hipótesis) puede ser una ecuación o ecuaciones, una distribución de probabilidad, una matriz de probabilidades, una matriz de pesos (en el caso de una red neuronal), etc.
<span></span><br><br>
    4. **<span style="color:red; font-size:18px">Predicción</span>**: ***Usando el modelo y nuevos datos*** (a priori nunca vistos por el algoritmo de aprendizaje) podemos ***realizar predicciones explotando el modelo***; es decir, aplicando las ecuaciones, distribuciones de probabilidad, matriz de probabilidades, etc.

### 2.2.2 Aprendizaje No Supervisado


* Para los ***algoritmos de aprendizaje no supervisado***, se necesitan ***datos no etiquetados***; es decir, que solo se conozca el estado (una prueba diagnostica, una transacción bancaria, etc) pero no el resultado. Por tanto solo se necesita saber el dominio de entrada de los datos.


* El ***proceso*** dentro del aprendizaje no supervisado sería el siguiente, tanto ***para la resolución de problemas de Clusterización*** (o agrupación) como de ***Reducción de la dimensionalidad***:
<span></span><br><br>
    1. **<span style="color:red; font-size:18px">Datos de Entrenamiento</span>**: Se necesitan un conjunto de datos de entrenamiento donde:
        + Se conozcan el dominio o las ***varibles de entrada 'X'***.
<span></span><br><br>
    2. **<span style="color:red; font-size:18px">Algoritmos de Aprendizaje</span>**: El algoritmo de aprendizaje es el ***encargado de general un modelo*** que posteriormente sirva para predecir. Este modelo lo genera ***minimizando*** algún tipo de ***función de coste*** que nos permita ***validar si el modelo generado en cada iteración es mejor o peor que en la iteración anterior***. Los algoritmos de aprendizaje no supervisado siguen el siguiente ***flujo iterativo***:
<span></span><br><br>  
        - ***Generación o Actualización del modelo***: El algoritmo ***genera un modelo*** basandose en las observaciones.
<span></span><br><br>
        - ***Evaluación del modelo***: Se ***evalua si el modelo es mejor o peor*** que el generado anteriormente ***viendo la función de coste***. Si es mejor se sigue entrenando el modelo y si es peor se deja de entrenar el modelo y nos quedamos como resultado con el mejor modelo.
<span></span><br><br>
    3. **<span style="color:red; font-size:18px">Modelo</span>**: Es una **"representación de la realidad"** que el ***algoritmo de aprendizaje ha aprendido basandose en los datos de entrenamiento***. 
<span></span><br><br>
    4. **<span style="color:red; font-size:18px">Predicción</span>**: ***Usando el modelo y nuevos datos*** (a priori nunca vistos por el algoritmo de aprendizaje) podemos ***tomar decisiones explotando el modelo***.

### 2.2.3 Aprendizaje por Refuerzo


* El aprendizaje por refuerzo esta basado en un ***AGENTE*** que **"vive"** en un ***ENTORNO*** y es capaz de percibir un ***ESTADO*** de ese entorno. El ***agente puede ejecutar ACCIONES en cada estado*** y estas acciones conllevan diferentes ***RECOMPENSAS*** y que pueden mover también al agente a otro estado del entorno.


* El objetivo del Aprendizaje por Refuerzo es aprender una ***POLÍTICA*** que es la ***encargada de tomar la decisión de qué acción tomar en cada estado*** con el objetivo de ***maximizar la recompensa***, bien sea a corto o largo plazo.


* Por hacer un símil con el aprendizaje supervisado, podemos considerar la ***Política*** como el ***Modelo*** del aprendizaje por refuerzo.


* El ***proceso*** dentro del aprendizaje por refuerzo sería el siguiente:
<span></span><br><br>
    1. **<span style="color:red; font-size:18px">Acción</span>**: El ***agente*** basansose en su ***estado*** y su ***política*** toma la decisión de realizar una ***acción***.
<span></span><br><br>
    2. **<span style="color:red; font-size:18px">Recompensa</span>**: Al realizar la ***acción*** el agente ***evalua*** su estado y recibe una ***recompensa o castigo***.
<span></span><br><br>
    3. **<span style="color:red; font-size:18px">Estado</span>**: Tras realizar la ***acción, el estado cambia***.
<span></span><br><br>
    4. **<span style="color:red; font-size:18px">Política</span>**: En función de la ***recompensa o castigo*** que recibe el agente por la ***acción tomada, modifica (o no) su política*** de tomas de decisiones.

## <a name="M3">3. Esquema general del (Deep) Machine Learning</a>


* El ***objetivo del (Deep) Machine Learning*** es el de ***crear sistemas que sean capaces de aprender por ellos mismo sin ser programados de forma explicita y aplicar lo aprendido en un entorno real para la predicción, toma de decisiones, etc.***


* ***Lo que aprenden estos sistemas se plasma en un modelo*** que luego tiene que ser ***explotado para la predicción de nuevos casos, acciones, etc.***


* Por tanto podemos decir que el objetivo final es el de ***crear el mejor modelo posible para poder explotarlo***.


* Para ***crear el mejor modelo posible debemos de seguir una metodología*** que podría ser la siguiente:
<span></span><br><br>
    1. **<span style="color:red; font-size:18px">Selección de Datos</span>**: Debemos de tener un conjunto de datos lo suficientemente grande para que un algoritmo de aprendizaje aprenda. Este conjunto de datos lo tenemos que dividir en: 
<span></span><br><br>
        + ***Conjunto de entrenamiento*** (train): Datos utilizados para ***entrenar el modelo***.
<span></span><br><br>
        + ***Conjunto de test***: Datos utilizados para ***evaluar el rendimiento del modelo***.
<span></span><br><br>
        * Existen diferentes ***técnicas para la selección de los datos de entrenamiento y test***, siendo estas técnicas de evaluación las siguientes:
<span></span><br><br>
            - ***Resustitución***: Todos los datos disponibles se utilizan como datos de test y de entrenamiento.
<span></span><br><br>
            - ***Partición (Hold Out)***: Divide los datos en dos subconjuntos: uno de entrenamiento y uno de test.
<span></span><br><br>
            - ***Validación cruzada (Cross Validation)***: Divide los datos aleatoriamente en ‘N’ bloques. Cada bloque se utiliza como test para un sistema entrenado por el resto de bloques.
<span></span><br><br>
            - ***Exclusión Individual (Leaving One Out)***: Este método utiliza cada dato individual como dato único de test de un sistema entrenado con todos los datos excepto el de test.
<span></span><br><br>
    2. **<span style="color:red; font-size:18px">Selección del Algoritmo de Aprendizaje</span>**: Hay que ***seleccionar un tipo u otro de algoritmo de aprendizaje*** en función del problema que se quiera resolver bien sea:
<span></span><br><br>
        - ***Regresión***
        - ***Clasificación***
        - ***Clustering***
        - ***Reducción de la dimensionalidad***
<span></span><br><br>
        * A modo de resumen ver algunas de las ***características*** que cumplen los ***problemas de Regresión, Clasificación y Clustering*** para seleccionar de manera adecuada el problema a resolver:
<span></span><br><br>
<img src="./imgs/13_09_tipos_algoritmos.png" style="width: 600px;"/>
<span></span><br><br>      
    3. **<span style="color:red; font-size:18px">Evaluación del Rendimiento del Modelo</span>** (Aprendizaje Supervisado): Otro de los problemas que tenemos que resolver es el de ***ver lo bueno o malo que es el modelo generado*** con el algoritmo de aprendizaje. Para ello tenemos una serie de ***métricas de evaluación*** que nos permitiran ***ver el nivel de acierto o la tasa de error cometidas en las predicciones*** bien sea ***con los datos de test*** (datos que no se han usado para el entrenamiento) o con los datos de entrenamiento (que son datos que el modelo ya ha visto y que por tanto ya conoce y debería predecirlos bien).
<span></span><br><br>
        * Sin entrar mucho en profundidad sobre las ***métricas de evaluación***, enumeramos las métricas más utilizadas para los problemas de regresión y clasificación:
<span></span><br><br>
            - ***Regresión:***
                + Error Absoluto Medio (MAE)
                + Error Cuadratico Medio (MSE)
<span></span><br><br>                
            - ***Clasificación:***
                + Accuracy
                + Precision
                + Recall
                + F1 (Precision-Recall)
<span></span><br><br>
    4. **<span style="color:red; font-size:18px">Explotación del Modelo</span>**: El último paso es el de explotar el modelo, ***realizando predicciones o tomando decisiones con nuevos datos***.

## <a name="M4">4. Overfiting vs Underfiting</a>


* Uno de los puntos más interesantes en el ML, es que se haya conseguido ***un aprendizaje correcto con los datos de entrenamiento***, obteniendo un modelo capaz de ***generalizar correctamente los nuevos datos de entrada***. 


* Por el contrario, el ***obtener un modelo como resultado de un sobreajuste (overfitting) o sobregeneralización (underfitting) de los datos de entrenamiento, hará que la salida proporcionada con nuevos datos de entrada tenga (muy probablemente) un error muy elevado, lo que quiere decir que la predicción no será correcta.***
