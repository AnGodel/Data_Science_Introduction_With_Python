* En primer lugar vamos a importar las librería necesarias:



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
%matplotlib inline 
```

<hr>


### 1. Carga de Datos


* Los datos se encuentran en el fichero: **"./data/precio_casas.csv"**

```python
df = pd.read_csv('../data/precio_casas.csv', header=0, sep=';')
df.head()
```

<hr>


### 2. Cambio de estructura de datos a numpy

```python
X = df[['pies_cuadrados_casa', 'num_banios', 'num_habitaciones', 'num_cocinas', 'num_coches_garaje', 'pies_cuadrados_garaje']].values
y = df['precio_dolares'].values
```

<hr>


### 3. Visualización de las relaciones entre variables

```python
sns.pairplot(df)
```

<hr>


### 4. Creacción del modelo y ajuste del mismo

```python
# Creamos un objeto de la clase LinearRegression
reg_mul = linear_model.LinearRegression()

# Entrenamos el modelo pasandole las variables independientes y dependiente
reg_mul.fit(X, y)
```

<hr>


### 5. Obtención del modelo

```python
# Obtenemos los coheficientes
betas = reg_mul.coef_
beta_0 = reg_mul.intercept_


# Siendo
# X1 = pies_cuadrados_casa
# X2 = num_banios
# X3 = num_habitaciones
# X4 = num_cocinas
# X5 = num_coches_garaje
# X6 = pies_cuadrados_garaje

print ("Y = {b0:0.2f} + {b1:0.2f}X1 + {b2:0.2f}X2 + {b3:0.2f}X3 + {b4:0.2f}X4 + {b5:0.2f}X5 + {b6:0.2f}X6"
       .format(b0=beta_0, b1=betas[0], b2=betas[1], b3=betas[2], b4=betas[3], b5=betas[4], b6=betas[5]))
```

<hr>


### 6. Medición de la calidad de los resultados: Obtención del Coeficiente de Determinación R<sup>2</sup>

```python
r2 = reg_mul.score(X, y)

print ("Coeficiente de determinación R^2: {r2}".format(r2=r2))
```
