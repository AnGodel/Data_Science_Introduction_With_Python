# 12 - Graficos Seaborn


* El siguiente Notebook tiene como objetivo ***mostrar*** algunos de los tipos de ***gráficos más importantes*** para la ***representación visual de información*** con la librería de ***Seaborn***.



* ***Seaborn*** es una librería basada en MatplotLib que facilita la creación de gráficos sobre todo cuando trabajamos con DataFrames de Pandas, ya que su representación es directa.


* Para más información de los gráficos que permite Seaborn, visitar el siguiente enlace: https://seaborn.pydata.org/api.html
    

* En esta Notebook mostraremos como realizar los siguientes ***gráficos con Seaborn***:

    1. Gráfico de Barras
    2. Gráfico de Lineas
    3. Gráfico de dispersión
    4. Histogramas
    5. Gráficos de cajas y bigotes
    6. Gráficos de Violines
    
    
* Hay otros gráficos interesantes que nos pueden aportar información más relevante sobre los datos:

    7. Pairplot: Ver las relaciones entre cada par de variables del Dataset.
    8. Joinplot: Muestra la relación entre dos pares de variables y sus respectivas distribuciones.
    9. Lmplot: Muestra la relación entre dos pares de variables y su recta de regresión.


<hr>


* Para mostrar los diferentes tipos de gráficos vamos a usar las siguientes fuentes de datos que nos facilíta la propia librería de Seaborn:

    - ***Tips***: Dataset con datos de comidas en restaurantes.
    - ***Titanic***: Dataset con los pasajeros embarcados en el Titanic.
    - ***Iris***: Dataset con el tamaño del petalo y sepalo de 3 tipos de flores.
    - ***Flights***: Registros de vuelos.


* En primer lugar vamos a importar la librería de Matplotlib y Seaborn (Necesitamos importar las dos):

```python
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns
```

### Importamos los datos

```python
df_tips = sns.load_dataset("tips")
df_titanic = sns.load_dataset("titanic")
df_iris = sns.load_dataset("iris")
df_flights = sns.load_dataset("flights")
```

```python
df_tips.sample(3)
```

```python
df_titanic.sample(3)
```

```python
df_iris.sample(3)
```

```python
df_flights.sample(3)
```

## 1. Gráfico de Barras


* Un gráfico de barras es una ***representación gráfica en un eje cartesiano de las frecuencias de una variable cualitativa o discreta***. 


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.barplot.html


* Ejemplo: Supervivientes del titanic por sexo

```python
sns.barplot(x="sex", y="survived", data=df_titanic)
```

* De forma muy sencilla podemos representar los supervivientes del titanic por sexo y clase en la que viajaban:

```python
sns.barplot(x="sex", y="survived", hue="class", data=df_titanic)
```

<hr>


## 2. Gráfico de Lineas


* Un gráfico de líneas es una ***representación gráfica en un eje cartesiano*** de la ***relación que existe entre dos variables reflejando con claridad los cambios producidos***.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.lineplot.html


* Ejemplo: Número de pasajeros por año

```python
sns.lineplot(x="year", y="passengers", data=df_flights)
```

* Podemos visualizarlo en diferentes series; por ejemplo, por año y mes:

```python
sns.lineplot(x="year", y="passengers", hue="month", data=df_flights)
```

<hr>


## 3. Gráfico de dispersión


* Un gráfico de dispersión muestra en un eje cartesiano la ***relación que existe entre dos variables***.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.scatterplot.html


* Ejemplo: Relación de tamaños del alto del petalo y sepalo de la flor

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=df_iris)
```

* Podemos diferenciarlas por tipo de Flor:

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df_iris)
```

<hr>


## 4. Histogramas


* Un histograma es un gráfico de barras vertical que ***representa la distribución de frecuencias*** de un conjunto de datos.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.kdeplot.html


* Ejemplo: Distribución de la factura.

```python
sns.kdeplot(df_tips['total_bill'], shade=True, color="g")
plt.title('Distribución del importe de las Facturas')
plt.xlabel('Impote')
plt.ylabel('%')
```

<hr>


## 5. Gráficos de cajas y bigotes


* Los diagramas de Caja-Bigotes (boxplots o box and whiskers) son una presentación visual que ***describe varias características importantes, al mismo tiempo, tales como la dispersión y simetría***.


* Para su realización se representan los ***tres cuartiles y los valores mínimo y máximo de los datos***, sobre un rectángulo, alineado horizontal o verticalmente.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.boxplot.html


* Ejemplo: Distribución de edades de los pasajeros del titanic por clase en la que viajaban.


```python
sns.boxplot(x="class", y="age", data=df_titanic)
```

* Podemos diferenciarlo también por sexo

```python
sns.boxplot(x="class", y="age", hue="sex", data=df_titanic)
```

<hr>


## 6. Gráficos de Violines

* Un diagrama de violín se utiliza para ***visualizar la distribución de los datos y su densidad de probabilidad***.


* Este gráfico es una combinación de un diagrama de cajas y bigotes y un diagrama de densidad girado y colocado a cada lado, para ***mostrar la forma de distribución de los datos***.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.violinplot.html


* Ejemplo: Distribución del importe de la factura.


```python
sns.violinplot(x="total_bill", data=df_tips)
```

* Podemos diferenciarlos por dia de la semana:

```python
sns.violinplot(x="day", y="total_bill", data=df_tips)
```

<hr>

## 7. Pairplot


* Este tipo de gráfico nos permite ver las relaciones entre cada par de variables del Dataset.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.pairplot.html


* Ejemplo: Relación entre pares de variables numéricas del dataset del tamaño de las flores.

```python
sns.pairplot(df_iris)
```

* Podemmos diferenciarlas por alguna variable; por ejemplo, por el tipo de flor

```python
sns.pairplot(df_iris, hue="species", palette="husl")
```

<hr>


## 8. Joinplot


* Muestra la relación entre dos pares de variables y sus respectivas distribuciones.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
sns.jointplot(x="total_bill", y="tip", data=df_tips)
```

<hr>


## 9. Lmplot

* Muestra la relación entre dos pares de variables y su recta de regresión.


* Para más información del gráfico en Seaborn: https://seaborn.pydata.org/generated/seaborn.lmplot.html

```python
sns.lmplot(x="total_bill", y="tip", data=df_tips)
```

* Podemos diferenciarlas por alguna clase.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df_tips)
```
