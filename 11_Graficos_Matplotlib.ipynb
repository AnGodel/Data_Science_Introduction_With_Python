## 1. Gráfico de Barras


* Un gráfico de barras es una ***representación gráfica en un eje cartesiano de las frecuencias de una variable cualitativa o discreta***. 


* La ***orientación*** del gráfico puede ser:

    - ***Vertical***: las distintas categorías están situadas en el eje horizontal y las barras de frecuencias crecen verticalmente.
    - ***Horizontal***: las categorías se sitúan en el eje vertical y las barras crecen horizontalmente. Suelen usarse cuando hay muchas categorías o sus nombres son demasiado largos.
    
    
* Documentación: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html


* Ejemplo: Representación de los goles marcados por los jugadores de la selección Española:

```python
futbolistas = { "Villa": 59, "Raúl": 44, "Torres": 38, "Silva": 35, 
               "Hierro": 29, "Morientes": 27, "Butragueño": 26, "D.Stéfano": 23, 
               "J.Salinas": 22, "Michel": 21, "Pirri":16, "Xb.Alonso": 16, 
               "Ramos": 16, "Santillana": 15, "Fàbregas": 15, "Cazorla": 14, 
               "Iniesta": 14, "Morata": 13, "Luis Enrique": 12}

x = list(futbolistas.keys()) # Paso las claves del diccionario futbolistas a una lista
y = list(futbolistas.values()) # Paso los vales del diccionario futbolistas a un lista
```

* Creamos la gráfica y pintamos los datos en Vertical:

```python
# Creamos la gráfica
fig, ax =  plt.subplots(figsize=(20, 6))

ax.bar(x, y, color = 'green')
ax.set_title('Goles Marcados por los miembros de la Selección Española')
plt.plot()
```

* Pintamos la gráfica en Horizontal:

```python
# Creamos la gráfica
fig, ax =  plt.subplots(figsize=(10, 10))

# Creamos las barra Horizontales (los goles)
plt.barh(np.arange(len(x)), y, color='red')
 
# Ponemos las etiquetas a las barras (los nombres)
plt.yticks(np.arange(len(x)), x)
plt.show()
```

## 2. Gráfico de Lineas


* Un gráfico de líneas es una ***representación gráfica en un eje cartesiano*** de la ***relación que existe entre dos variables reflejando con claridad los cambios producidos***.


* Se suelen usar para presentar tendencias temporales.


* Ejemplo sencillo:

```python
data = np.random.randn(20) # genera 20 número aleatorios

# Creamos la gráfica
fig, ax =  plt.subplots(figsize=(8, 5))

ax.plot(data, color = 'blue')
plt.plot()
```

* Ejemplo con multiples series

```python
serie1 = 1 - np.random.randn(20) # genera 20 número aleatorios
serie2 = 2 + np.random.randn(20) # genera 20 número aleatorios
serie3 = 4 + np.random.randn(20) # genera 20 número aleatorios

# Creamos la gráfica
fig, ax =  plt.subplots(figsize=(15, 6))

# Pintamos cada una de las series
ax.plot(serie1, color = 'red')
ax.plot(serie2, color = 'green')
ax.plot(serie3, color = 'blue')

# Añadimos Formato al gráfico
ax.set_title('Ejemplo de gráfico de Lineas')
ax.set_xlabel('Numero de elemento de la serie')
ax.set_ylabel('Valor del elemento')
ax.grid(True)
plt.plot()
```

## 3. Gráfico de dispersión


* Un gráfico de dispersión muestra en un eje cartesiano la ***relación que existe entre dos variables***.





```python
x = np.random.randn(100)
y = np.random.randn(100)

# Creamos la gráfica de dispersion
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'b+')
ax.plot(2*y, 2*x, 'r.')
plt.show()
```

## 4. Gráfico de sectores


* Un gráfico de sectores es una ***representación circular de las frecuencias relativas de una variable cualitativa o discreta que permite, de una manera sencilla y rápida, su comparación***.

```python
labels = 'Gana Local (1)', 'Empate (X)', 'Gana Visitante (2)'
resultados = [179, 115, 86]

# Creamos el pie chart
fig, ax = plt.subplots(figsize=(5, 5))
ax.pie(resultados, labels=labels, autopct='%1.1f%%')

# Damos formato al gráfico
ax.set_title('Resultados Quiniela 2017-18')
plt.plot()
```

## 5. Histogramas


* Un histograma es un gráfico de barras vertical que ***representa la distribución de frecuencias*** de un conjunto de datos.

```python
# Generamos una distribución normal de media 176 y de desviación típica 7.6
media = 176
desviacion = 7.6

# Generamos 100K puntos
x = media + desviacion * np.random.randn(100000)

# Creamos el histograma
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(x, bins=100, color='g')

# Damos formato al gráfico
ax.set_xlabel('Altura en cm')
ax.set_ylabel('Ocurrencias por bins')
ax.set_title('Distribución de la Altura de los Españoles')
ax.grid(True)
plt.show()
```

## 7. Gráficos de cajas y bigotes


* Los diagramas de Caja-Bigotes (boxplots o box and whiskers) son una presentación visual que ***describe varias características importantes, al mismo tiempo, tales como la dispersión y simetría***.


* Para su realización se representan los ***tres cuartiles y los valores mínimo y máximo de los datos***, sobre un rectángulo, alineado horizontal o verticalmente.


```python
# Generamos una distribución normal de media 166 y de desviación típica 10
alt_mujeres = np.random.normal(166, 8.2, 100) 
# Generamos una distribución normal de media 176 y de desviación típica 7.6
alt_hombres = np.random.normal(176, 7.6, 100) 

fig, ax = plt.subplots(figsize=(8, 6))
plt.boxplot([alt_mujeres, alt_hombres], labels= ['Mujeres', 'Hombres'])
ax.set_ylabel('Altura (cm)')
plt.show()
```

<hr>


## 8. Gráficos de Violines


* Un diagrama de violín se utiliza para ***visualizar la distribución de los datos y su densidad de probabilidad***.


* Este gráfico es una combinación de un diagrama de cajas y bigotes y un diagrama de densidad girado y colocado a cada lado, para ***mostrar la forma de distribución de los datos***.

```python
# Generamos una distribución normal de media 166 y de desviación típica 10
alt_mujeres = 166 + 10 * np.random.randn(100) 
# Generamos una distribución normal de media 176 y de desviación típica 7.6
alt_hombres = 176 + 7.6 * np.random.randn(100) 

fig, ax = plt.subplots(figsize=(8, 6))

# plot violin plot
ax.violinplot([alt_mujeres, alt_hombres], showmeans=False, showmedians=True)
ax.set_title('Altura Mujeres y Hombres')
plt.show()
```
